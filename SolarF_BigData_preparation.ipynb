```python
def get_XY_by_reading_csv(dir_path):
    tempX = []
    tempY = []

    for dirname, _, filenames in os.walk(dir_path):
        print('Total data file: ' + str(len(filenames)))
        #print(filenames[1423])
        mcount = 0

        for i, filename in enumerate(filenames):
       
            tmp_df = pd.read_csv(os.path.join(dirname, filename), delimiter= '\t')
            tmp_df = tmp_df.iloc[:, 1:25]
            #tmp_df = impute_missing_value(tmp_df)
            if not tmp_df.isnull().values.any():
                tempX.append(tmp_df.T)
                tempY.append(filename[0]) #Flare class M,X,...
            else:
                mcount += 1
        print('#files with missing value: ' + str(mcount))
            
    return np.array(tempX), np.array(tempY)
```

```python
def get_partition_data(partition_location):
    
    FL_data_path = partition_location + '/FL'
    NF_data_path = partition_location + '/NF'
    
    
    X, Y = get_XY_by_reading_csv(FL_data_path)
    Xnf, Ynf = get_XY_by_reading_csv(NF_data_path)
    Xcomb = np.vstack((X , Xnf)) 
    Ycomb = np.concatenate((Y , Ynf))
    
    return Xcomb, Ycomb
```
